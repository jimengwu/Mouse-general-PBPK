```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.dpi"] = 300
plt.rcParams["figure.figsize"] = (10,4)
```

```python
def check_non_empty_cols(df):
    return [df[col].name for col in df if df[col].any()]
```

```python
liver_data = pd.read_excel("invitro_human_liver.xlsx")
liver_data = liver_data[:255]
```

```python
liver_data.head()
```

```python
year = liver_data[["Study_ID","Year of publication"]].drop_duplicates()["Year of publication"].value_counts().to_frame()
```

```python
year = year.sort_index()
```

```python
# %matplotlib notebook
```

```python

ax = year.plot(kind='bar',alpha=0.5,edgecolor='black', linewidth=1.2)
# ax.grid()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
substance = liver_data[["Study_ID","Substance name"]].drop_duplicates()["Substance name"].value_counts().to_frame()
# substance
```

```python

ax = substance.plot(kind='barh',alpha=0.5,edgecolor='black', linewidth=1.5,color='#3c6e71')
ax.set_yticklabels(substance.index,wrap=True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
# figure(figsize=(10, 6), dpi=80)
substance_core = liver_data[["Study_ID","Substance core"]].drop_duplicates()["Substance core"].value_counts().to_frame()
```

```python
ax = substance_core.plot(kind='barh',alpha=0.5,edgecolor='black', linewidth=1.2,color='#e3655b')
ax.set_yticklabels(substance_core.index,wrap=True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

# On one kind of materials

```python
study_Ag = liver_data[liver_data["Substance name"] == "Ag"]
study_Ag
```

```python
study_Ag[check_non_empty_cols(study_Ag)].columns
```

```python
results_study_Ag = study_Ag[['Study_ID','Substance name','In vitro concentration','Genotoxicity: cytokinesis-block micronucleus assay [total micronucleus number]',
       'Genotoxicity: alkaline comet assay (DNA damage) [Tail Moment]',
       'Cytotoxicity: MTT assay [viability % of control]',
       'Cytotoxicity: neutral red uptake (NRU) assays  [viability % of control]',
       'Cytotoxicity: cell titer blue (CTB)  [viability % of control]',
       'Cytotoxicity: DAPI assay [% viability of control]',
       'Cytotoxicity: alamar blue assay [viability % of control]']]
results_study_Ag
```

```python
results_study_Ag.to_csv("test.csv")
```

```python
results_study_Ag["results"] = results_study_Ag[['Genotoxicity: cytokinesis-block micronucleus assay [total micronucleus number]',
       'Genotoxicity: alkaline comet assay (DNA damage) [Tail Moment]',
       'Cytotoxicity: MTT assay [viability % of control]',
       'Cytotoxicity: neutral red uptake (NRU) assays  [viability % of control]',
       'Cytotoxicity: cell titer blue (CTB)  [viability % of control]',
       'Cytotoxicity: DAPI assay [% viability of control]',
       'Cytotoxicity: alamar blue assay [viability % of control]']].sum(axis=1)
results_study_Ag
```
